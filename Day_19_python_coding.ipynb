```python

students = []
classes = ("Math", "Science", "English", "History")

def show_menu():
    print("\n=== STUDENT MANAGEMENT SYSTEM ===")
    print("1. Add Student")
    print("2. View All Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Add Grades")
    print("7. View Performance")
    print("8. Exit")
    print("================================")

def add_student():
    print("\n--- Add New Student ---")

    student_id = input("Enter Student ID: ")
    name = input("Enter Student Name: ")

    # Check if student already exists
    for student in students:
        if student['id'] == student_id:
            print("Error: Student ID already exists!")
            return

    # Get age with error handling
    try:
        age = int(input("Enter Age: "))
    except:
        print("Error: Please enter a valid number for age!")
        return

    print("Available classes:", classes)
    student_class = input("Enter Class: ")

    # Check if class is valid
    if student_class not in classes:
        print("Error: Invalid class!")
        return

    # Create student dictionary
    student = {'id': student_id,'name': name,'age': age,'class': student_class,'grades': []}

    students.append(student)
    print(f"Student {name} added successfully!")

def view_all_students():
    print("\n--- All Students ---")

    if len(students) == 0:
        print("No students found!")
        return

    for student in students:
        print(f"ID: {student['id']}, Name: {student['name']}, Age: {student['age']}, Class: {student['class']}")

def search_student():
    print("\n--- Search Student ---")

    search_id = input("Enter Student ID to search: ")

    for student in students:
        if student['id'] == search_id:
            print("\nStudent Found:")
            print(f"ID: {student['id']}")
            print(f"Name: {student['name']}")
            print(f"Age: {student['age']}")
            print(f"Class: {student['class']}")
            if student['grades']:
                print(f"Grades: {student['grades']}")
            else:
                print("Grades: No grades yet")
            return

    print("Student not found!")

def update_student():
    print("\n--- Update Student ---")

    search_id = input("Enter Student ID to update: ")

    for student in students:
        if student['id'] == search_id:
            print(f"\nCurrent Info: {student['name']}, Age: {student['age']}, Class: {student['class']}")

            # Update name
            new_name = input("Enter new name (press enter to keep current): ")
            if new_name:
                student['name'] = new_name

            # Update age
            new_age = input("Enter new age (press enter to keep current): ")
            if new_age:
                try:
                    student['age'] = int(new_age)
                except:
                    print("Invalid age! Keeping current age.")

            # Update class
            new_class = input("Enter new class (press enter to keep current): ")
            if new_class:
                if new_class in classes:
                    student['class'] = new_class
                else:
                    print("Invalid class! Keeping current class.")

            print("Student updated successfully!")
            return

    print("Student not found!")

def delete_student():
    print("\n--- Delete Student ---")

    search_id = input("Enter Student ID to delete: ")

    for i in range(len(students)):
        if students[i]['id'] == search_id:
            deleted_name = students[i]['name']
            students.pop(i)
            print(f"Student {deleted_name} deleted successfully!")
            return

    print("Student not found!")

def add_grades():
    print("\n--- Add Grades ---")

    search_id = input("Enter Student ID: ")

    for student in students:
        if student['id'] == search_id:
            try:
                grade = float(input("Enter grade (0-100): "))

                if grade < 0 or grade > 100:
                    print("Error: Grade must be between 0-100!")
                    return

                student['grades'].append(grade)
                print(f"Grade {grade} added for {student['name']}!")
            except:
                print("Error: Please enter a valid number!")
            return

    print("Student not found!")

def check_performance(grade):
    if grade >= 90:
        return "Excellent"
    elif grade >= 80:
        return "Very Good"
    elif grade >= 70:
        return "Good"
    elif grade >= 60:
        return "Average"
    else:
        return "Needs Improvement"

def view_performance():
    print("\n--- Student Performance ---")

    if len(students) == 0:
        print("No students found!")
        return

    for student in students:
        print(f"\nStudent: {student['name']} (ID: {student['id']})")

        if student['grades']:
            total = sum(student['grades'])
            average = total / len(student['grades'])
            performance = check_performance(average)

            print(f"Grades: {student['grades']}")
            print(f"Average: {average:.1f} - {performance}")
        else:
            print("No grades available")

# Main program
print("Welcome to Simple Student Management System!")

# Main loop
while True:
    show_menu()

    choice = input("Enter your choice (1-8): ")

    if choice == '1':
        add_student()
    elif choice == '2':
        view_all_students()
    elif choice == '3':
        search_student()
    elif choice == '4':
        update_student()
    elif choice == '5':
        delete_student()
    elif choice == '6':
        add_grades()
    elif choice == '7':
        view_performance()
    elif choice == '8':
        print("Thank you for using the system! Goodbye!")
        break
    else:
        print("Invalid choice! Please enter 1-8.")
```
