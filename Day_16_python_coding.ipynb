```python
import random

# 1. Generate scores
scores = [random.randint(0, 100) for i in range(10)]
print("Scores:", scores)

# 2. Display with index
for index,i in enumerate(scores):
    print(index,"--",i)

# 3. Scores above 75
high_scores = [i for i in scores if i > 75]
print("Scores greater than 75:", high_scores)

# 4. Average score
avg = sum(scores) / len(scores)
print("Average Score:", avg)
```
